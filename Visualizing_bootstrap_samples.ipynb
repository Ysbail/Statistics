```python
import numpy as np
import matplotlib.pyplot as plt
import pickle as pkl
from stat_func import ecdf
```

```python
with open('rainfall.pkl', 'rb') as f:
    rainfall = pkl.load(f)
```

### Generate bootstrap sample and plot ECDF from bootstrap sample

```python
for _ in range(50):
    bs_sample = np.random.choice(rainfall, size=len(rainfall))
    x, y = ecdf(bs_sample)
    plt.plot(x, y, marker='.', linestyle='none',
                 color='gray', alpha=0.1)
```

```python
x, y = ecdf(rainfall)
plt.plot(x, y, marker='.')
plt.margins(0.02)
plt.xlabel('yearly rainfall (mm)')
plt.ylabel('ECDF')
```
